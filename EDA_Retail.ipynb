# **The weak areas where one can work to make more profits are :**
**Sales and profit are more in Standard Class, so we must encourage it.**

**We should limit sales of furniture and increase that of technology and office suppliers as furniture has very less profit as compared to sales.**

**Considering the sub-categories, sales of tables should be minimized to enhance productivity.**

**Copiers have the highest number of Sales and Profits while tables have the lowest.**

**We should concentrate more on the states like California, Texas and New York to generate more profit as the highest ratio of sales and profits are from these areas.** 

**Sales and Profit are the least in Southern and Central region. So, we should give more incentives like discount in order to increase sales, eventually profit will also increase**.

**Focusing on few states like North Dakota, Wyoming and West Virginia is also required as these states have less or no sales at all.**

**To get maximum profits, Discount should be less than or equal to 50%.**
